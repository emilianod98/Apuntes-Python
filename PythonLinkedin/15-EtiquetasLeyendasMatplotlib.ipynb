```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/aeropuerto.csv", nrows= 100000)
data = np.unique(df.DayOfWeek, return_counts= True)
labs = ["Lun", "Mar", "Mie", "Jue", "Vie", "Sab", "Dom"]
data
```

```python
plt.pie(x= data[1],
        labels= data[0],
        radius= 1.5,
        colors= ["Red", "Orange", "Yellow", "Green", "Blue", "Gray", "Black"],
        startangle= 90)
plt.show()
```

```python
plt.pie(x= data[1],
        labels= labs,
        radius= 1.5,
        colors= ["Red", "Orange", "Yellow", "Green", "Blue", "Gray", "Black"],
        startangle= 90)
plt.show()
```

```python
plt.pie(x= data[1],
        labels= labs,
        radius= 1.5,
        colors= sns.color_palette("hls",7),
        startangle= 90)
plt.show()
```

```python
plt.pie(x= data[1],
        labels= labs,
        radius= 1.5,
        colors= sns.color_palette("hls",7),
        explode= (0, 0, 0, 0.2, 0, 0, 0.1),
        startangle= 90)
plt.show()
```

```python
plt.pie(x= data[1],
        labels= labs,
        radius= 1.5,
        colors= sns.color_palette("hls",7),
        explode= (0, 0, 0, 0.2, 0, 0, 0.1),
        startangle= 90,
        autopct= "%1.1f%%")
plt.show()
```

```python
plt.pie(x= data[1],
        radius= .6,
        colors= sns.color_palette("hls",7),
        explode= (0, 0, 0, 0.2, 0, 0, 0.1),
        startangle= 90,
        autopct= "%1.1f%%")
plt.legend(loc = "center left", labels = labs)
plt.show()
```

```python
plt = sns.barplot(x= labs, y= data[1])
plt.set(xlabel = "Da de la Semana", ylabel = "Nmero de Vuelos")
```
